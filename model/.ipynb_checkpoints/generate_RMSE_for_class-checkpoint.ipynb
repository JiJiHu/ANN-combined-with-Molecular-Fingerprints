```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
os.listdir()
```

```python
ds = pd.read_csv('G2_train_data - group devation.csv')
```

```python
ds.head()
```

```python
set(ds['class'])
```

```python
len(set(ds['class']))
```

```python
ds_class = pd.DataFrame(list(set(ds['class'])), columns=['Class'])
```

```python
ds_class
```

```python
a = []
b = []
for i in set(ds['class']):
    temp = ds[ds['class'] == i]
    n = len(temp)
    rmse = sqrt(mean_squared_error(temp.LogKOH, temp.PLogKOH))
    a.append(rmse)
    b.append(n)
```

```python
ds_class['RMSE'] = a
ds_class['Num'] = b
```

```python
ds_class
```

```python
ds_class.to_csv('class_RMSE.csv', index = False)
```
